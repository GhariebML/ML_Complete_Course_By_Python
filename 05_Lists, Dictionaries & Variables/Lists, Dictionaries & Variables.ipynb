```python
ThisAge = input('How old are you ? ')
ThisAge  = int(ThisAge)
ThisGender = input('What is your gender ? ')




if ThisAge <=10 : 
  if ThisGender =='m' : 
    print('child boy')
  elif ThisGender =='f' : 
    print('child girl')
elif ThisAge <18 : 
  if ThisGender =='m' : 
    print('Teen boy')
  elif ThisGender =='f' : 
    print('Teen girl')
elif ThisAge<= 40 : 
  if ThisGender =='m' : 
    print('Adult man')
  elif ThisGender =='f' : 
    print('Adult woman')
else : 
  if ThisGender =='m' : 
    print('old man')
  elif ThisGender =='f' : 
    print('old woman')
```
