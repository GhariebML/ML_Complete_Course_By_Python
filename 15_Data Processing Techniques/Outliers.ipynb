```python
x =  'Perimeter'

q75,q25 = np.percentile(data.loc[:,x],[75,25])
q75,q25 
```

```python
intr_qr = q75-q25
intr_qr
```

```python
Max = q75+(3*intr_qr)
Min = q25-(3*intr_qr)
Max,Min
```

```python
data.loc[data[x] < Min,x] = np.nan
data.loc[data[x] > Max,x] = np.nan

data = data.dropna(axis = 0)

data.reset_index(inplace=True)
data.drop(['index'],axis = 1,inplace=True)
```

```python
data['Perimeter'].mean(),data['Perimeter'].std()
```

```python
threshold = 3
outlier = []
for i in data['Perimeter'].tolist():
    z = (i-data['Perimeter'].mean())/data['Perimeter'].std()
#     print(z)
    if abs(z) > threshold:
        outlier.append(i)
print('outlier in dataset is', outlier)

```

```python
OutlierIndices = [i for i,j in enumerate(data['Perimeter'].tolist()) if j in outlier]
len(OutlierIndices)
```

```python
Data = data.drop(OutlierIndices,axis=0)
Data
```

```python

def estimate_gaussian(dataset):

    mu = np.mean(dataset)
    sigma = np.std(dataset)
    limit = sigma * 2

    min_threshold = mu - limit
    max_threshold = mu + limit

    return mu, sigma, min_threshold, max_threshold

mu, sigma, min_threshold, max_threshold = estimate_gaussian(data['Perimeter'].values)
mu, sigma, min_threshold, max_threshold 
```

```python
data[data['Perimeter']<min_threshold].shape
```

```python
data[data['Perimeter']>max_threshold].shape
```

```python
data.drop(data[data['Perimeter']>max_threshold].index,axis=0,inplace=True)
data.reset_index(inplace=True)
data.drop(['index'],axis = 1,inplace=True)
data
```
