```python
import pandas as pd
```

```python
data = pd.read_csv('Airline_Delay_Cause.csv')
```

```python
data
```

```python
data['arr_flights'].min(),data['arr_flights'].max()
```

```python
def Classes(x) : 
    if x <=70 : 
        return 'A'
    elif x < 100 : 
        return 'B'
    else : 
        return 'C'
        

data['ArrClasses'] = data['arr_flights'].apply(lambda x : Classes(x))
data['ArrClasses'].value_counts()
```

```python
data.isna().sum()
```

```python
data.dropna(inplace = True)
data.shape
```

```python
data.isna().sum()
```

```python
data.head()
```

```python
data.drop(['year', 'month', 'carrier', 'carrier_name', 'airport', 'airport_name'],axis=1,inplace=True)
data
```

```python
X = data.drop(['ArrClasses'],axis=1)
y = data['ArrClasses']
X
```

```python
from sklearn.feature_selection import SelectPercentile, chi2 , f_classif ,SelectKBest
```

```python
FeatureSelection = SelectPercentile(score_func = f_classif, percentile=50)
```

```python
X_Selected = FeatureSelection.fit_transform(X, y)
pd.DataFrame(X_Selected,columns =[i for i,j in zip(X.columns,FeatureSelection.get_support()) if j ] )
```

```python
def FeatureSelectionByPercentile(X,y,Percentile) : 
    FeatureSelection = SelectPercentile(score_func = f_classif, percentile=Percentile)
    X_Selected = FeatureSelection.fit_transform(X, y)
    NewData = pd.DataFrame(X_Selected,columns =[i for i,j in zip(X.columns,FeatureSelection.get_support()) if j ] )
    NewData['y'] = y
    return NewData
    
    
NewData = FeatureSelectionByPercentile(X,y,20)      
NewData.head()
```

```python
NewData = FeatureSelectionByPercentile(X,y,30)      
NewData.head()
```

```python
NewData = FeatureSelectionByPercentile(X,y,80)      
NewData.head()
```

```python
def FeatureSelectionByNumber(X,y,Number) : 
    FeatureSelection = SelectKBest(score_func = f_classif, k=Number)
    X_Selected = FeatureSelection.fit_transform(X, y)
    NewData = pd.DataFrame(X_Selected,columns =[i for i,j in zip(X.columns,FeatureSelection.get_support()) if j ] )
    NewData['y'] = y
    return NewData
    
    
NewData = FeatureSelectionByNumber(X,y,7)      
NewData.head()
```

```python
NewData = FeatureSelectionByNumber(X,y,3)      
NewData.head()
```
