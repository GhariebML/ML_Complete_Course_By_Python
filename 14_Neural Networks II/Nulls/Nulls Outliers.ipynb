```python
# 2 - Interquartile range(IQR)

data.boxplot('carrier_ct')
```

```python
import numpy as np
for x in ['carrier_ct']:
    q75,q25 = np.percentile(data.loc[:,x],[75,25])
    intr_qr = q75-q25
    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)
 
    data.loc[data[x] < min,x] = np.nan
    data.loc[data[x] > max,x] = np.nan

Data = data.dropna(axis = 0)

Data.reset_index(inplace=True)
Data.drop(['index'],axis = 1,inplace=True)
Data
```

```python
Data.describe()
```

```python

# 3 - Z Score

data.boxplot('carrier_ct')
data['carrier_ct'].mean(),data['carrier_ct'].std()
```

```python
threshold = 1
outlier = []
M,S = data['carrier_ct'].mean(),data['carrier_ct'].std()
for i in data['carrier_ct']:
    z = (i-(M))/S
    # print(z)
    if abs(z) > threshold:
        outlier.append(i)
print('outlier in dataset is', len(outlier))
```

```python
OutlierIndices = [i for i,j in enumerate(data['carrier_ct'].tolist()) if j in outlier]
OutlierIndices

data.loc[OutlierIndices,:]

Data = data.drop(OutlierIndices,axis=0)
Data.reset_index(inplace=True)
Data.drop(['index'],axis = 1,inplace=True)
Data
```

```python
Data.describe()
```

```python

# 4- Gaussian Distribution


def estimate_gaussian(dataset):

    mu = np.mean(dataset)
    sigma = np.std(dataset)
    limit = sigma * 1.5

    min_threshold = mu - limit
    max_threshold = mu + limit

    return mu, sigma, min_threshold, max_threshold

mu, sigma, min_threshold, max_threshold = estimate_gaussian(data['carrier_ct'].values)
mu, sigma, min_threshold, max_threshold 

data[data['carrier_ct']<min_threshold]



data[data['carrier_ct']>max_threshold]



data[data['carrier_ct']>max_threshold*2]



data.drop(data[data['carrier_ct']>max_threshold*2].index,axis=0,inplace=True)
data.reset_index(inplace=True)
data.drop(['index'],axis = 1,inplace=True)
data.describe()
```
