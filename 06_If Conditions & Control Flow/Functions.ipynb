```python
def TextAnalysis(text) : 
  def CharacterCount(MyText,Char) : 
    return str(MyText).count(Char)
  def OriginalWordLengthDef(Word) : 
    return len(Word)
  def CleanedWordLengthDef(Word) : 
    for char in '~!@#$%^&*()_+?' : 
      Word = Word.replace(char,'')
    return len(Word)

  SpecialCharDict = {}
  OriginalWordLengthDict = {}
  CleanedWordLengthDict = {}
  for i in '~!@#$%^&*()_+?' : 
    SpecialCharDict[i] = CharacterCount(text,i)
  for Word in text.split() : 
    OriginalWordLengthDict[Word] = OriginalWordLengthDef(Word)
    CleanedWordLengthDict[Word] = CleanedWordLengthDef(Word)

  
  return SpecialCharDict, OriginalWordLengthDict,CleanedWordLengthDict


```

```python
Phrase = 'hi!! how?? are$%  you &  my )brother!! '

TextAnalysis(Phrase)
```

```python
def fibon(n):
    a = b = 1
    result =[]
    for i in range(n):
        result.append (a)
        a , b = b , a+b
    return result

print (fibon(20))
```

```python
class Account:
    def __init__(self, name, account_number, initial_amount):
        self.name = name
        self.no = account_number
        self.balance = initial_amount
    def deposit(self, amount):
        self.balance += amount
    def withdraw(self, amount):
        self.balance -= amount
    def show(self):
        s = f'{self.name} with account number {self.no} has banalce of {self.balance} '
        print (s)
```

```python
a1 = Account('John Olsson', '19371554951', 20000)
a2 = Account('Liz Olsson', '19371564761', 50000)
```

```python
print("a1 balance :  " , a1.balance )
print("a2 no      :  " ,   a2.no)
```

```python

a1.deposit(1000)
a1.withdraw(4000)
a2.withdraw(10500)
a1.withdraw(3500)
```

```python
print ("a1's balance:", a1.balance)
print ("a2's balance:", a2.balance)
```

```python
a1.show()
a2.show()
```

```python
class Person:
    def __init__(self, name,mobile_phone=None, office_phone=None,private_phone=None, email=None):
        self.name = name
        self.mobile = mobile_phone
        self.office = office_phone
        self.private = private_phone
        self.email = email
    def add_mobile_phone(self, number):
        self.mobile = number
    def add_office_phone(self, number):
        self.office = number
    def add_private_phone(self, number):
        self.private = number
    def add_email(self, address):
        self.email = address
    def dump(self):
        s = self.name + '\n'
        if self.mobile is not None:
            s += 'mobile phone: %s\n' % self.mobile
        if self.office is not None:
            s += 'office phone: %s\n' % self.office
        if self.private is not None:
            s += 'private phone: %s\n' % self.private
        if self.email is not None:
            s += 'email address: %s\n' % self.email
        print (s)

p1 = Person('Hans Hanson',office_phone='767828283', email='h@hanshanson.com')
p2 = Person('Ole Olsen', office_phone='767828292')
p2.add_email('olsen@somemail.net')
phone_book = [p1, p2]

for person in phone_book:
    person.dump()
```

```python
class Degrees : 
  def __init__(self,Name,Subject,Degree) : 
    self.MyName = Name
    self.MySubject = Subject
    self.MyDegree = Degree

  def Percentage(self,Total) : 
    print(f'My name is {self.MyName} and my percentage is {100* self.MyDegree/Total}%')


  def ShowDetails(self) : 
    print(f'my name is {self.MyName} , I study {self.MySubject} and I got {self.MyDegree}')

```

```python
Student1 = Degrees('Mahmoud','Physics',20)
Student1.ShowDetails()
```
