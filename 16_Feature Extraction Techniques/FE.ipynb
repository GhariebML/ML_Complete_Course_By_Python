```python
%%time
data['TweetYear'] = data['created_at'].apply(lambda x : x.split()[5])
data['TweetMonth'] = data['created_at'].apply(lambda x : x.split()[1])
data['TweetDay'] = data['created_at'].apply(lambda x : x.split()[2])
# data['DayName'] = data['created_at'].apply(lambda x : x.split()[0])
data['WholeDate'] = data.apply(lambda x : pd.Timestamp(str(x['TweetYear']) + '-' + str(x['TweetMonth']) + '-' + str(x['TweetDay'])).day_name()  ,axis=1 )
data['TweetHour'] = data['created_at'].apply(lambda x : x.split()[3].split(':')[0])

def GetDayPartFromHour(n) : 
    n = int(n)
    if n<= 6 : 
        return 'Early Morning'
    elif  n <=10 :
        return 'Morning'
    elif n <= 13 : 
        return 'Noon'
    elif n <= 17 : 
        return 'Evening'
    else : 
        return 'Night'



data['DayPart'] = data['TweetHour'].apply( lambda ka3bosha : GetDayPartFromHour(ka3bosha))


# 'Wed Jan 07 11:14:08 +0000 2015'


def GetDayNameFromDate(h) : 
    ThisYear = h.split()[5]
    ThisMonth = h.split()[1]
    ThisDay= h.split()[2]
    WholeDate = str(ThisYear) + '-' + str(ThisMonth) + '-' + str(ThisDay)
    DayName = pd.Timestamp(WholeDate).day_name()
    return DayName

data['NameOfDay'] = data['created_at'].apply(lambda x : GetDayNameFromDate(x))


def CalculateDayNameFromDate(Y,M,D) : 

    WholeDate = str(Y) + '-' + str(M) + '-' + str(D)
    DayName = pd.Timestamp(WholeDate).day_name()
    return DayName
    



data['NewNameOfDay']  = data.apply(lambda x : CalculateDayNameFromDate(x['TweetYear'],x['TweetMonth'],x['TweetDay']) ,axis=1 )
```
