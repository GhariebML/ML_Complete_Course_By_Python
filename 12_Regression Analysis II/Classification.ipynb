```python
import pandas as pd
```

```python
data = pd.read_csv('Airline_Delay_Cause.csv')
```

```python
data
```

```python
data.dropna(inplace = True)
```

```python
data.columns
```

```python
y = data['carrier']
X = data[['arr_flights', 'arr_del15', 'carrier_ct', 'weather_ct', 'nas_ct',
       'security_ct', 'late_aircraft_ct', 'arr_cancelled', 'arr_diverted',
       'arr_delay', 'carrier_delay', 'weather_delay', 'nas_delay',
       'security_delay', 'late_aircraft_delay']]
```

```python
X.info()
```

```python
y.value_counts()
```

```python
Data = data[data['carrier'].isin(['OO','DL'])]
Data
```

```python
Data['carrier'].value_counts()
```

```python
y = Data['carrier']
X = Data[['arr_flights', 'arr_del15', 'carrier_ct', 'weather_ct', 'nas_ct',
       'security_ct', 'late_aircraft_ct', 'arr_cancelled', 'arr_diverted',
       'arr_delay', 'carrier_delay', 'weather_delay', 'nas_delay',
       'security_delay', 'late_aircraft_delay']]
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=44, shuffle =True)

#Splitted Data
print('X_train shape is ' , X_train.shape)
print('X_test shape is ' , X_test.shape)
print('y_train shape is ' , y_train.shape)
print('y_test shape is ' , y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RandomForestClassifierModel = RandomForestClassifier(criterion = 'gini',n_estimators=100,max_depth=2,random_state=33)
```

```python
RandomForestClassifierModel.fit(X_train, y_train)
```

```python
y_pred = RandomForestClassifierModel.predict(X_test)
y_pred
```

```python
from sklearn.metrics import classification_report
```

```python
ClassificationReport = classification_report(y_test,y_pred)
print('Classification Report is : \n', ClassificationReport )
```

```python
#Import Libraries
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
#----------------------------------------------------

#----------------------------------------------------
#Calculating Confusion Matrix
CM = confusion_matrix(y_test, y_pred)
#print('Confusion Matrix is : \n', CM)

# drawing confusion matrix
sns.heatmap(CM, center = True,cmap='Blues_r')
plt.show()
```
