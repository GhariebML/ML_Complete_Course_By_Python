```python
AllDF = pd.DataFrame()

AllCSVFiles = []
SubFolderNumber = {}
LabelsColumns,JPColumns,Ratio = [],[],[]
NullsInLabel,JPLabel= [],[]
Path = r'C:\Users\hesha\Desktop\App2\data_new'
for F in tqdm(os.listdir(Path)) : 
    SubFolder = os.path.join(Path,F)
    if os.path.isdir(SubFolder)  : 
        SubFolderNumber[F] = len(os.listdir(SubFolder))
        for SmallFolder in os.listdir(SubFolder) : 
            FinalFolder = os.path.join(SubFolder,SmallFolder)
            if os.path.isdir(FinalFolder) : 
                for File in os.listdir(FinalFolder) : 
                    AllCSVFiles.append(File)
                    LabelCSVFile = os.path.join(FinalFolder,'Labels.csv')
                    JPCSVFile = os.path.join(FinalFolder,'Joint_Positions.csv')
                    LabelFile = pd.read_csv(LabelCSVFile, header=None)
                    NullsInLabel.append(list(LabelFile.isna().sum().values)[0])
                    JPFile = pd.read_csv(JPCSVFile, header=None)
                    JPLabel.append(list(JPFile.isna().sum().values)[0])
                    JPRowsNumber = JPFile.shape[0]
                    FinalRowsNumber = int(JPRowsNumber/25)
                    AllRows = []
                    for r in range(FinalRowsNumber) : 
                        ThisRow = []
                        for i in range(25) : 
                            ThisRow.extend(list(JPFile.loc[i + (25*r),:].values))
                        AllRows.append(ThisRow)
                    JPDataFrame = pd.DataFrame(AllRows) 
                    
                    JPDataFrame['MainFolder'] = F
                    JPDataFrame['SubFolder'] = SmallFolder

                    JPDataFrame['Labels'] = LabelFile[0]
                    AllDF = pd.concat([AllDF,JPDataFrame],axis=0)
#                     print(AllDF.shape)
                    
                    
```
